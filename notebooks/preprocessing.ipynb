```python
import os
from pydub import AudioSegment
import noisereduce as nr
import librosa
import soundfile as sf
```

```python
audio_dir = '/content/drive/MyDrive/audiocorpus'
output_dir = '/content/drive/MyDrive/Audio_Processed'
```

```python
def preprocess_audio(file_path):
    audio, sr = librosa.load(file_path, sr=16000)
    reduced_noise = nr.reduce_noise(y=audio, sr=sr)
    return reduced_noise, sr

for filename in os.listdir(audio_dir):
    if filename.endswith('.mp3'):
        file_path = os.path.join(audio_dir, filename)
        processed_audio, sr = preprocess_audio(file_path)
        # Save the processed audio
        sf.write(f'{output_dir}processed_{filename}', processed_audio, sr)
```
